```python
from decimal import Decimal, ROUND_UP

# Assuming 'products' is your DataFrame

# Convert the column to string first to avoid the error
products["Unit Cost USD"] = products["Unit Cost USD"].astype(str)

# Remove dollar sign
products["Unit Cost USD"] = products["Unit Cost USD"].str.replace("$", "", regex=False)

# Convert to float and handle errors (if any)
products["Unit Cost USD"] = pd.to_numeric(products["Unit Cost USD"], errors='coerce')

# Convert to Decimal and round up to 2 decimal places
products['Unit Cost USD'] = products['Unit Cost USD'].apply(
    lambda x: Decimal(x).quantize(Decimal('0.00'), rounding=ROUND_UP) if pd.notna(x) else Decimal(0)
)

# Fill NaN values (if any remain) with Decimal(0)
products["Unit Cost USD"].fillna(Decimal(0).quantize(Decimal('0.00')), inplace=True)
```

```python
# Convert the column to string first to avoid the error
products["Unit Price USD"] = products["Unit Price USD"].astype(str)

# Remove dollar sign
products["Unit Price USD"] = products["Unit Price USD"].str.replace("$", "", regex=False)

# Convert to float and handle errors (if any)
products["Unit Price USD"] = pd.to_numeric(products["Unit Price USD"], errors='coerce')

# Convert to Decimal and round up to 2 decimal places
products['Unit Price USD'] = products['Unit Price USD'].apply(
    lambda x: Decimal(x).quantize(Decimal('0.00'), rounding=ROUND_UP) if pd.notna(x) else Decimal(0)
)

# Fill NaN values (if any remain) with Decimal(0)
products["Unit Price USD"].fillna(Decimal(0).quantize(Decimal('0.00')), inplace=True)
```

```python
#Data Cleaning

#Define primary and unique key columns

primary_keys_sales = ['Order Number', 'Line Item', 'CustomerKey', 'StoreKey', 'ProductKey']
```

```python
#handling missing values
sales.dropna(subset=['Order Date'], inplace = True)
```

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'],errors = 'coerce')
```

```python
#Handling outliers
sales['Quantity'] = np.where(sales['Quantity'] > sales['Quantity'].quantile(0.99),sales['Quantity'].quantile(0.99),sales['Quantity'])
```

```python
#Removing duplicates
cleaned_sales = sales.drop_duplicates()
```

```python
#High percentage of missing values
threshold = 0.5
columns_to_drop = [col for col in cleaned_sales.columns if cleaned_sales[col].isnull().sum() / len(cleaned_sales) > threshold and col not in primary_keys_sales]
```

```python
# Drop irrelevant, low variance, redundant, high correlation, and privacy-sensitive columns as per conditions
cleaned_sales.drop(columns=set(columns_to_drop), inplace=True)
```

```python
cleaned_sales.to_csv("Downloads\cleaned_sales.csv",index = False)
```

# Loading cleaned datasets

```python
customers = pd.read_csv("Downloads/Customers_cleaned.csv",encoding = 'ISO 8859-1')
sales = pd.read_csv("Downloads/cleaned_sales.csv",encoding = 'ISO 8859-1')
stores = pd.read_csv("Downloads/cleaned_stores.csv",encoding = 'ISO 8859-1')
products = pd.read_csv("Downloads/Cleaned_products.csv", encoding = 'ISO 8859-1')
exchange_rates = pd.read_csv("Downloads/cleaned_exchange_rates.csv",encoding = 'ISO 8859-1')
```

```python
#merging sales with customers
sales_customers = pd.merge(sales, customers, how='left', left_on='CustomerKey', right_on='Customer_ID')
```

```python
# Merge sales_customers with stores
sales_customers_stores = pd.merge(sales_customers, stores, how='left', left_on='StoreKey', right_on='StoreKey')
```

```python
# Merge sales_customers_stores with products
sales_customers_stores_products = pd.merge(sales_customers_stores, products, how='left', left_on='ProductKey', right_on='ProductKey')
```

```python
# Save the merged dataset
sales_customers_stores_products.to_csv('Downloads\merged_data.csv', index=False)

print("Data merging completed.")
```
